```python
def get_flux_prompt(image_url): 
    input = {
        "image": image_url,
        "prompt": '''Can you give me a flux prompt to create this image? 
        Ensure that the prompt follows the following format:
        Example Prompt format: A hyperrealistic portrait of a weathered sailor in his 60s, with deep-set blue eyes, a salt-and-pepper beard, and sun-weathered skin. He is wearing a faded blue captains hat and a thick wool sweater. The background shows a misty harbor at dawn, with fishing boats barely visible in the distance.
        A well-crafted FLUX.1 prompt typically includes the following components: 
        Subject: The main focus of the image.
        Style: The artistic approach or visual aesthetic.
        Composition: How elements are arranged within the frame.
        Lighting: The type and quality of light in the scene.
        Color Palette: The dominant colors or color scheme.
        Mood/Atmosphere: The emotional tone or ambiance of the image.
        Technical Details: Camera settings, perspective, or specific visual techniques.
        Additional Elements: Supporting details or background information.'''
    }
    out = replicate.run(
        "yorickvp/llava-13b:80537f9eead1a5bfa72d5ac6ea6414379be41d4d4f6679fd776e9535d1eb58bb",
        input=input
    )
    out_list = ""
    for item in out:
        out_list += item
    return out_list

output = get_flux_prompt("https://r2.photoai.com/1734853448-1734851574-09f270247cd9316078f2020681dcc457.png")
print(output)
```

```python
# Cell 1: Imports and Configuration
import boto3
import requests
from motor.motor_asyncio import AsyncIOMotorClient
import asyncio
from datetime import datetime
import os
from dotenv import load_dotenv
load_dotenv('/Users/aayushchaudhary/Desktop/vscode/whatif/backend/.env')  # Specify the path to your .env file

# S3 Configuration
aws_access_key_id = os.getenv('AWS_ACCESS_KEY_ID')
aws_secret_access_key = os.getenv('AWS_SECRET_ACCESS_KEY')
s3_client = boto3.client('s3', region_name='us-east-1')  # Replace 'your-region' with your S3 region

BUCKET_NAME = "whatif-genai"
MONGODB_URI = os.getenv('MONGODB_URI')

mongo_client = AsyncIOMotorClient(MONGODB_URI, tlsAllowInvalidCertificates=True)
db = mongo_client.whatif
feed_collection = db.feed
```

```python
# Cell 3: Helper Functions
def download_image(url):
    """Download image from URL"""
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    return None

def upload_to_s3(image_data, object_name):
    """Upload image to S3 bucket"""
    try:
        s3_client.put_object(
            Bucket=BUCKET_NAME,
            Key=f"examples/{object_name}",
            Body=image_data,
            ContentType='image/png'
        )
        return f"https://{BUCKET_NAME}.s3.amazonaws.com/examples/{object_name}"
    except Exception as e:
        print(f"Error uploading to S3: {e}")
        return None

async def store_in_mongodb(image_data):
    """Store image data in MongoDB"""
    try:
        image_data["created_at"] = datetime.utcnow()
        image_data["type"] = "example"
        await feed_collection.insert_one(image_data)
        return True
    except Exception as e:
        print(f"Error storing in MongoDB: {e}")
        return False
```

```python
import requests
# Cell 4: Processing Function
async def process_examples():
    """Process all example images"""
    for example in image_prompts:
        # Download image from r2.photoai.com
        print(f"Processing {example['image_url']}")
        image_data = download_image(example['image_url'])
        if not image_data:
            print(f"Failed to download image from {example['image_url']}")
            continue
            
        # Generate S3 object name from original URL
        object_name = example['image_url'].split('/')[-1]
        
        # Upload to S3
        s3_url = upload_to_s3(image_data, object_name)
        if not s3_url:
            print(f"Failed to upload image to S3")
            continue
            
        # Store in MongoDB
        mongo_data = {
            
            "original_url": example['image_url'],
            "s3_url": s3_url,
            "prompt": example['prompt']
        }
        
        if await store_in_mongodb(mongo_data):
            print(f"Successfully processed {example['prompt']}")
        else:
            print(f"Failed to store in MongoDB for {example['prompt']}")
# Cell 5: Run Processing
await process_examples()
print("Processing complete!")
```
